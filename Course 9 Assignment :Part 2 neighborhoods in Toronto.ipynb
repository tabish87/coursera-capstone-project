```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
!pip install lxml
```

```python
url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
```

```python
soup = BeautifulSoup(url,'lxml')
```

```python
match =soup.find('table')
print(match)
```

```python
headers_list =[]
list_list=[]
```

```python
for trow in match.tbody.find_all('tr'):
    for theader in trow.find_all('th'):
        headers_list.append((theader).text)
```

```python
headers_list
```

```python
for trow in match.tbody.find_all('tr'):
    for tdata in trow.find_all('td'):
        list_list.append((tdata).text)
```

```python
list_list
```

```python
Canada = pd.DataFrame(np.array(list_list).reshape(289,3))
```

```python
Canada
```

```python
Canada[2] = Canada[2].map(lambda x: x.rstrip('\n'))
```

```python
Canada
```

```python
df_headers = pd.DataFrame(np.array(headers_list).reshape(1,3))
df_headers[2] = df_headers[2].map(lambda x: x.rstrip('\n'))
```

```python
df_headers
```

```python
Canada.rename(columns={0:df_headers[0][0]}, inplace=True)
Canada.rename(columns={1:df_headers[1][0]}, inplace=True)
Canada.rename(columns={2:df_headers[2][0]}, inplace=True)
```

```python
Canada
```

```python
Canada_without = Canada[Canada.Borough != "Not assigned"]
```

```python
Canada_without=Canada_without.reset_index(drop=True)
```

```python
Canada_without
```

```python

Neighbourhood = [Canada_without['Neighbourhood'][0]]  
Postcode = [Canada_without['Postcode'][0]]
Borough = [Canada_without['Borough'][0]]


for a, b, c in zip(Canada_without['Neighbourhood'][1:], Canada_without['Postcode'][1:], Canada_without['Borough'][1:]):
    if b == Postcode[-1]:        # If b matches the last value in result2,
        Neighbourhood[-1] += " , " + a  # add a to the last value of result1
    else:  # Otherwise add a new row with the values
        Neighbourhood.append(a)
        Postcode.append(b)
        Borough.append(c)
 
        
# Create a new dataframe using these result lists
summary = pd.DataFrame({df_headers[0][0]: Postcode, df_headers[1][0]: Borough, df_headers[2][0]: Neighbourhood })

summary = summary[[df_headers[0][0], df_headers[1][0], df_headers[2][0]]]

summary
```

```python
CanadaM= summary
CanadaM.loc[CanadaM['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = CanadaM['Borough']
CanadaM.head(12)
```

```python
CanadaM.shape
```

```python
maps = 'http://cocl.us/Geospatial_data'
latlong = pd.read_csv(maps)
latlong
```

```python

CanadaM_longlat= CanadaM.merge(latlong, how='inner', left_on='Postcode', right_on='Postal Code')
CanadaM_longlat
del CanadaM_longlat['Postal Code']
```

```python
CanadaM_longlat.rename(columns={"Postcode":'Postal Code'}, inplace = True)
```
