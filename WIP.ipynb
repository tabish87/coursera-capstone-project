# Toronto Postal Codes 

```python
toronto = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
soup = BeautifulSoup(toronto,'lxml')
match =soup.find('table')
print(match)
```

```python
headers_list =[]
list_list=[]
```

```python
for trow in match.tbody.find_all('tr'):
    for theader in trow.find_all('th'):
        headers_list.append((theader).text)
```

```python
headers_list
```

```python
for trow in match.tbody.find_all('tr'):
    for tdata in trow.find_all('td'):
        list_list.append((tdata).text)
```

```python
list_list
```

```python
toronto_df= pd.DataFrame(np.array(list_list).reshape(289,3))
```

```python
toronto_df
```

```python
toronto_df[2] = toronto_df[2].map(lambda x: x.rstrip('\n'))
```

```python
toronto_df
```

```python
df_headers = pd.DataFrame(np.array(headers_list).reshape(1,3))
df_headers[1] = df_headers[1].map(lambda x: x.rstrip('\n'))
df_headers[2] = df_headers[2].map(lambda x: x.rstrip('\n'))
df_headers[0] = df_headers[0].map(lambda x: x.rstrip('\n'))
```

```python
df_headers
```

```python
toronto_df.rename(columns={0:df_headers[0][0]}, inplace=True)
toronto_df.rename(columns={1:df_headers[1][0]}, inplace=True)
toronto_df.rename(columns={2:df_headers[2][0]}, inplace=True)
```

```python
toronto_df
```

```python
toronto_df_without = toronto_df[toronto_df.Borough != "Not assigned"]
```

```python
toronto_df_without=toronto_df_without.reset_index(drop=True)
toronto_df_without 
```

```python
Neighbourhood = [toronto_df_without['Neighbourhood'][0]]  
Postcode = [toronto_df_without['Postcode'][0]]
Borough = [toronto_df_without['Borough'][0]]


for a, b, c in zip(toronto_df_without['Neighbourhood'][1:], toronto_df_without['Postcode'][1:], toronto_df_without['Borough'][1:]):
    if b == Postcode[-1]:        # If b matches the last value in result2,
        Neighbourhood[-1] += " , " + a  # add a to the last value of result1
    else:  # Otherwise add a new row with the values
        Neighbourhood.append(a)
        Postcode.append(b)
        Borough.append(c)
 
        
# Create a new dataframe using these result lists
summary = pd.DataFrame({df_headers[0][0]: Postcode, df_headers[1][0]: Borough, df_headers[2][0]: Neighbourhood })

summary = summary[[df_headers[0][0], df_headers[1][0], df_headers[2][0]]]

summary
```

```python
toronto_final= summary
toronto_final.loc[toronto_final['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = toronto_final['Borough']
toronto_final.head(12)
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url3 = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url3).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```
